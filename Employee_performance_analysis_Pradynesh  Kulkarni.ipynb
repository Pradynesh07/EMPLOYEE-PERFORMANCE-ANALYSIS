### Imputing Outliers

```python
# 'data' is our DataFrame
columns_to_process = ['TotalWorkExperienceInYears','TrainingTimesLastYear','ExperienceYearsAtThisCompany','ExperienceYearsInCurrentRole',
                      'YearsSinceLastPromotion','YearsWithCurrManager']

for column in columns_to_process:
    # Calculate Interquantile range
    IQR = data[column].quantile(0.75) - data[column].quantile(0.25)

    # Calculate lower and upper bridges
    lower_quartile = data[column].quantile(0.25) - (1.5 * IQR)
    
    upper_quartile = data[column].quantile(0.75) + (1.5 * IQR)


    # Replace values less than lower quartile with lower quartile value
    data.loc[data[column] <= lower_quartile, column] = lower_quartile
    
    # Replace values greater than upper quartile with upper quartile value
    data.loc[data[column] >= upper_quartile, column] = upper_quartile
```

#### Checking Outliers After Imputing

### HyperParameter Tunning of Decision Tree using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Initialize the Decision Tree model
tree_clf = DecisionTreeClassifier(random_state=42)

# Define the hyperparameter grid to search
params = {
    "criterion":("gini", "entropy"), #used to measure the purity of split
    "splitter":("best", "random"), # searches the features for a split
    "max_depth":(list(range(1, 20))), #depth/height of tree range from 1 to 19
    "min_samples_split":[2, 3, 4],    #the minimum number of samples required to split internal node
    "min_samples_leaf":list(range(1, 20)),#minimum number of samples required to be at a leaf node,
    #we are passing list which is range from 1 to 19
}

# Initialize GridSearchCV
grid_search = GridSearchCV(tree_clf, params, cv=3, scoring='accuracy', n_jobs=-1,verbose=1)
#passing model to gridsearchCV ,
#tree_clf-->model
#params---->hyperparametes(dictionary we created)
#scoring--->performance matrix to check performance
#n_jobs---->Number of jobs to run in parallel,-1 means using all processors.
#verbose=Controls the verbosity: the higher, the more messages.it is used to print messages
#>1 : the computation time for each fold and parameter candidate is displayed;
#>2 : the score is also displayed;
#>3 : the fold and candidate parameter indexes are also displayed together with the starting time of the computation.
#cv------> number of flods/cross validation


# Fit the model with the best hyperparameters
grid_search.fit(X_train, y_train)

# Get the best model
best_Decision_tree_model = grid_search.best_estimator_

# Make predictions on the training set
decision_tree_train_predictions = best_Decision_tree_model.predict(X_train)

# Evaluate the Random Forest model on training set
train_accuracy = accuracy_score(y_train, decision_tree_train_predictions)
print(f"Training Accuracy: {train_accuracy*100:.2f}")

# Make predictions on the test set
decision_tree_test_predictions = best_Decision_tree_model.predict(X_test)

# Evaluate the Decision Tree model on test set
test_accuracy = accuracy_score(y_test, decision_tree_test_predictions)
print(f"Test Accuracy: {test_accuracy*100:.2f}")

# Print the best hyperparameters found
print("Best Hyperparameters:")
print(grid_search.best_params_)

# Print other metrics such as classification report or confusion matrix for the test set
print("\nClassification Report (Test Set):")
print(classification_report(y_test, decision_tree_test_predictions))

print("\nConfusion Matrix (Test Set):")
print(confusion_matrix(y_test, decision_tree_test_predictions))
```

**After Use of Hyperparameter, the accuracy of the model increases.**
